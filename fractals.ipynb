```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Fractal:
    def __init__(self, max_magnitude=4, max_iteration=120,n_points=1000,x1=-2.25,x2=0.75,y1=-1.5,y2=1.5):
        self.n_points = n_points
        
        # Create points in our sample, this will be the atlas rectangle
        self.xvalues = np.linspace(x1, x2, self.n_points)
        self.yvalues = np.linspace(y1, y2, self.n_points)
        
        # size of these lists of x and y values
        self.x_len = len(self.xvalues)
        self.y_len = len(self.yvalues)
        

        #madnitud is the distance fromt he center to the point
        self.max_magnitude=max_magnitude
        
        #The max number of iterations, if the achieve this number means that the function didn't diverge and will stop, if not we can continue until the infinite
        self.max_iteration=max_iteration
        
        self.atlas = np.zeros((self.x_len,self.y_len))

    #this method will create the array with the points to be draw in out final step
    def create_atlas_mandelbrot(self):
        self.atlas = np.zeros((self.x_len,self.y_len))
        for ix in range(self.x_len):
            for iy in range(self.y_len):
                
                #create c, this is a complex number
                cx = self.xvalues[ix]
                cy = self.yvalues[iy]
                c = complex(cx, cy)
                
                # now we know what c is for this place in the atlas, apply the mandel() function to return the number of iterations it took to diverge
                self.atlas[ix,iy] = self.mandelbrot_calculate(c)

    # this method get the mx number of iteration, more means more detail and procesing time, c must be fixed
    def mandelbrot_calculate(self,c):
        # z is a complex number that start with a value of 0+0i, this value will be iterated
        z = complex(0,0)
    
        # start iterating and stop when it's done maxiter times
        for iteration in range(self.max_iteration):
            z = (z*z) + c
        
            # if the magnitude is bigger than the max magnitud setted, and if so stop iterating as we've diverged already
            if abs(z) > self.max_magnitude:
                break
        
        # as there is a match between each point in our attlas and our complex numbers the colors will be set by the iteration when it diverge
        return iteration

    #this method will create the array with the points to be draw in out final step
    def create_atlas_julia(self,cx,cy):

        c = complex(cx, cy) 
        self.atlas = np.zeros((self.x_len,self.y_len))
        for ix in range(self.x_len):
            for iy in range(self.y_len):
                
                #create c, this is a complex number
                zx = self.xvalues[ix]
                zy = self.yvalues[iy]
                z = complex(zx, zy)
                
                # now we know what c is for this place in the atlas, apply the mandel() function to return the number of iterations it took to diverge
                self.atlas[ix,iy] = self.julia_calculate(z,c)

    # this method get the mx number of iteration, more means more detail and procesing time, c must be fixed
    def julia_calculate(self,z,c):
        # start iterating and stop when it's done maxiter times
        for iteration in range(self.max_iteration):
            z = (z*z) + c
        
            # if the magnitude is bigger than the max magnitud setted, and if so stop iterating as we've diverged already
            if abs(z) > self.max_magnitude:
                break
        
        # as there is a match between each point in our attlas and our complex numbers the colors will be set by the iteration when it diverge
        return iteration



    # Update points in our sample, this will be the atlas rectangle
    def update_sample(self, x1,x2,y1,y2):
        # Create points in our sample, this will be the atlas rectangle
        self.xvalues = np.linspace(x1, x2, self.n_points)
        self.yvalues = np.linspace(y1, y2, self.n_points)
        
        # size of these lists of x and y values
        self.x_len = len(self.xvalues)
        self.y_len = len(self.yvalues)

    #it is like main, the names by now can be "mandelbrot" or "julia", cx and cy are just for julia
    def draw(self,fractal_name='mandelbrot',cx=-0.768662,cy=0.0930477):   
        if fractal_name=='mandelbrot':
            self.create_atlas_mandelbrot()
        elif fractal_name=='julia':
            self.create_atlas_julia(cx,cy)
        else:
            print("YOU MULST USE mandelbrot or julia")
            return
        # figure size
        plt.figure(figsize=(18,18))
        
        # we need traspose because is puthon
        plt.imshow(self.atlas.T, interpolation="nearest", cmap="jet")
```
